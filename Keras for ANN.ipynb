```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
type(iris)
```

```python
print(iris.DESCR)
```

```python
X = iris.data
X
```

```python
y = iris.target
y
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_obj = MinMaxScaler()
scaler_obj.fit(X_train)
```

```python
scaled_X_train = scaler_obj.transform(X_train)
```

```python
scaled_X_test = scaler_obj.transform(X_test)
```

```python
type(scaled_X_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(scaled_X_train,y_train,epochs=500, verbose=2)
```

```python
prediction_classes = np.argmax(model.predict(scaled_X_test), axis=1)
prediction_classes
```

```python
model.metrics_names
```

```python
model.evaluate(x=scaled_X_test,y=y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test.argmax(axis=1), prediction_classes)
```

```python
print(classification_report(y_test.argmax(axis=1), prediction_classes))
```
